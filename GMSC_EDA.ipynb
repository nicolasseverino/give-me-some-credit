## Visualization plots

```python
df$SeriousDlqin2yrs <- factor(df$SeriousDlqin2yrs)
df <- na.omit(df)
```

```python
plotBoxPlot <- function(feature) {
    ggplot(na.omit( df ), aes_string(x="SeriousDlqin2yrs", y=feature)) + 
    geom_boxplot(notch=FALSE)
    # ggtitle("Age density plots for defaulted and non-defaulted borrowers")
} 

#plotBoxPlot("age")
```

```python
for (i in 1:length(colnames(df))) {
    feature = colnames(df)[i]
    print ( plotBoxPlot(feature) )
}
```

```python
plotFeatureDensity <- function(feature, outliers=FALSE) {
    
    if (outliers) {    
        x_min <- min(df[,feature])
        x_max <- max(df[,feature])
    }
    
    else {
        if (min(df[,feature]) < 0.5*quantile(df[,feature], 0.1) ) {
            x_min <- 0.5*quantile(df[,feature], 0.1)
        }
        else {
            x_min <- min(df[,feature])
        }
        
        if (max(df[,feature]) > 2*quantile(df[,feature], 0.9) ) {
            x_max <- 2*quantile(df[,feature], 0.9)
        }
        else {
            x_max <- max(df[,feature])
        }
        
    }
    
    ggplot(df, aes_string(x=feature, color="SeriousDlqin2yrs")) + 
    geom_density() + 
    scale_color_discrete(name = "", breaks=c("0", "1"), labels=c("Did Not Default", "Defaulted"))  +
    ggtitle(paste(feature, "density plots for defaulted and non-defaulted borrowers")) + 
    theme(legend.position="top") +
    xlim(x_min, x_max)

}

#plotFeatureDensity("age", outliers=TRUE)
#plotFeatureDensity("age", outliers=FALSE)
```
